```python
rent_df.dropna() #결측치가 포함된 행들 삭제
```

```python
rent_df.dropna(subset = ['Size'])  #Size에 있는 결측치 행들만 삭제
```

```python
rent_df.dropna(1)  # Rent 앞뒤에 열 2개 삭제(BHK, Size)/ 1은 axis 1 의미한다. 
```

```python
rent_df.drop(['BHK', 'Size'], axis = 1)  ## BHK, Size 컬럼 날리기-->근데 보통 컬럼을 날리진 않는다. 
#                                            이런 경우는 의미 없는 행의 결측치가 너무 많은 경우만 해당됨.
```

```python
rent_df[rent_df['Size'].isna()]   #Size에 있는 결측치만 추려봄
na_index =rent_df[rent_df['Size'].isna()].index   #이걸  index로 추려서 뽑아서 변수에 저장.
```

```python
rent_df.iloc[na_index]    # 바로 위 행의 결측치만 출력
```

```python
rent_df.fillna(-99)   # 결측치 채우기
```

```python
rent_df.fillna(-99).loc[na_index ] # 채워진 결측치만 출력
```

```python
rent_df.fillna(rent_df.median()).loc[na_index]  # 결측치에 평균 값을 넣어준다. mean을 넣을수도 있고, median을 넣을수도 있다. 
```

```python
rent_df = rent_df.fillna(rent_df.median())   # 채워진 결측치를 원본에 복사
```

```python
rent_df.isna().mean()   # 결측치 남아 있는지 재확인!
```
